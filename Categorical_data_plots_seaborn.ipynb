```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure your DataFrame df exists with the correct structure and columns.
plt.figure(figsize=(10, 10))

# Use size instead of s and ensure it maps directly to a column in the DataFrame
sns.scatterplot(
    x='Sales',
    y='Discount',
    size='Profit',  # Use 'Profit' for size mapping
    hue='Category',
    data=df,
    sizes=(100, 1000)  # Set the range of sizes for markers
)

plt.title("Sales vs Discount by Category with Profit as Size")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure your DataFrame df exists with the correct structure and columns.
plt.figure(figsize=(10, 10))

# Use 'size' for marker sizes and map it to a column in the DataFrame
sns.scatterplot(
    x="Sales",
    y="Profit",
    size="Shipping Cost",  # Use 'Shipping Cost' as the size variable
    hue="Market",          # Color markers by the 'Market' column
    data=df,
    sizes=(20, 200)        # Adjust the range of marker sizes
)

plt.title("Sales vs Profit by Market with Shipping Cost as Size")
plt.show()
```
